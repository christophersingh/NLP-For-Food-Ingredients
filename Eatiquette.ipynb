```python
import math
import re
import nltk
from string import punctuation
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
from nltk.tokenize import MWETokenizer
from nltk.probability import FreqDist
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.collocations import *
from nltk import bigrams, trigrams
from nltk.collocations import BigramCollocationFinder
import collections
#from nltk.metrics import BigramAssocMeasures, BigramCollocationFinder

str_articles = str(clean_text)

punctuations = '!"#$%&\'()*+,-{\\[\\]}.â€”/:;<=>?@[\\]^_`{|}~'
stop_words = set(stopwords.words('english'))

                      
vocabulary = sorted(set(clean_text))        
lexical_div = len(words_text) / len(vocabulary)

freq_words = {}
for token in clean_text:
        if token in freq_words.keys():
            freq_words[token] += 1
        else:
            freq_words[token] = 1

        
words_fd = nltk.FreqDist(clean_text)
words_fd_most = words_fd.most_common(75)

#Determine the top words in the dataset
def most_frequent(freq, n):
    result = {}
    index = 0
    for word in sorted(freq, key=freq.get, reverse=True):
        index += 1
        result[word] = freq[word]
        if index == n: 
            break
    return result

#Apply collocations to see which words are commonly used with each other 
def text_collocations(text, n):
    bigrams = nltk.collocations.BigramAssocMeasures()
    finder = BigramCollocationFinder.from_words(text)
    finder.apply_freq_filter(n)
    collocations = finder.nbest(bigrams.pmi, 20)
    return collocations

#Summarize the text after data cleaning
def text_summary(text, n):
    sentence = sent_tokenize(text.lower())
    sentences = []
    for sent in sentence:
        if sent not in punctuations:
            sentences.append(sent)
    
    w_tokens = nltk.word_tokenize(text.lower())
    tokens = []
    for token in w_tokens:
        if token not in punctuations:
            if token not in stop_words:
                token = lemmatizer.lemmatize(token)
                tokens.append(token)
                clean_text = nltk.Text(tokens)
                collocations = text_collocations(clean_text ,n)

    summary = []
    for c in collocations:
        for sentence in sentences:
            term1 = c[0]
            term2 = c[1]
            term = c[0] + ' ' + c[1]
            if (term in sentence):
                if (sentence not in summary):
                    summary.append(sentence)
                
    return summary

print('\033[1mTotal number of vocabulary is\033[0m {}'.format(len(vocabulary)))
print()
top_words_no = 100
top_words = most_frequent(freq_words, top_words_no)
print('\033[1mTop {} frequent words:\033[0m'.format(top_words_no))
top_w = []
for i in top_words:
    top_w.append(i)
print(top_w)
```

```python
#Identify topics that are found in LDA

import warnings
warnings.simplefilter("ignore", DeprecationWarning)
# Load the LDA model from sk-learn
from sklearn.decomposition import LatentDirichletAllocation as LDA
 
# Helper function for vectorization
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
#I set the topics to 5. Adding more than 10 topics results in longer wait times
number_topics = 5
number_words = 10
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
# Print the topics found by the LDA model
print("Topics found via LDA:")
print_topics(lda, count_vectorizer, number_words)
```
